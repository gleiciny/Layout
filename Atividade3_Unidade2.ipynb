```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Criando o DataFrame
dados = {
    'Dia': ['Segunda', 'Tera', 'Quarta', 'Quinta', 'Sexta', 'Sbado', 'Domingo'],
    'Vendas': [1200, 1500, 1800, 1700, 2200, 2500, 2100],
    'Clientes': [50, 60, 70, 65, 80, 90, 85],
    'Lucro': [300, 400, 450, 420, 550, 600, 510]
}

df = pd.DataFrame(dados)

# Garantindo que 'Dia' seja tratada como varivel categrica com a ordem correta
df['Dia'] = pd.Categorical(df['Dia'], categories=['Segunda', 'Tera', 'Quarta', 'Quinta', 'Sexta', 'Sbado', 'Domingo'], ordered=True)

# Grfico de barras para o total de vendas por dia, associando 'Dia' ao 'hue' para a paleta
plt.figure(figsize=(12, 7))
sns.barplot(x='Dia', y='Vendas', data=df, hue='Dia', palette='Set2')  # Agora associa 'Dia' ao 'hue'
plt.title('Total de Vendas por Dia', fontsize=16, fontweight='bold')
plt.xlabel('Dia da Semana', fontsize=14)
plt.ylabel('Vendas (em R$)', fontsize=14)
plt.xticks(rotation=45, fontsize=12)  # Rotacionando as labels dos dias
plt.yticks(fontsize=12)
plt.tight_layout()  # Ajuste o layout para no cortar rtulos
plt.show()


# Grfico de disperso (scatter plot) entre o nmero de clientes e o total de vendas
plt.figure(figsize=(12, 7))
sns.scatterplot(x='Clientes', y='Vendas', data=df, hue='Dia', palette='Set2', s=150, marker='o', edgecolor='black', linewidth=1.5)
plt.title('Nmero de Clientes vs Total de Vendas', fontsize=16, fontweight='bold')
plt.xlabel('Nmero de Clientes', fontsize=14)
plt.ylabel('Vendas (em R$)', fontsize=14)
plt.legend(title='Dia da Semana', loc='upper left', fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()
plt.show()

# Corrigindo o clculo da correlao e o heatmap
correlacao = df[['Vendas', 'Clientes', 'Lucro']].corr()

# Grfico do Heatmap para mostrar a correlao entre as variveis
plt.figure(figsize=(10, 8))
sns.heatmap(correlacao, annot=True, cmap='Blues', fmt='.2f', linewidths=1.5, square=True, cbar_kws={'label': 'Correlao'})
plt.title('Correlao entre Vendas, Clientes e Lucro', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```
